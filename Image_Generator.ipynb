```python
from huggingface_hub import create_repo, upload_file, HfApi
import time

# Replace with your desired space name
space_name = "ai-image-generator-by-aniket"
api = HfApi()


# Upload files
upload_file(path_or_fileobj="app.py", path_in_repo="app.py", repo_id=space_name, repo_type="space")
upload_file(path_or_fileobj="requirements.txt", path_in_repo="requirements.txt", repo_id=space_name, repo_type="space")
```

```python
# Create app.py
with open("app.py", "w") as f:
    f.write("""
import gradio as gr
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
).to("cuda")

def generate(prompt):
    image = pipe(prompt).images[0]
    return image

gr.Interface(fn=generate,
             inputs=gr.Textbox(label="Enter prompt"),
             outputs="image",
             title="AI Image Generator").launch()
""")

# Create requirements.txt
with open("requirements.txt", "w") as f:
    f.write("gradio\ndiffusers\ntransformers\ntorch\nsafetensors\naccelerate")
```
